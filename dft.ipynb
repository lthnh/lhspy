```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
f = 40.0
fs = 100.0
ts = 1.0 / fs
n = np.array(range(0, 65))
print('The fundamental frequency: ', fs / len(n))
sample = np.sin(2.0 * np.pi * f * n * ts)
plt.plot(n, sample, 'o')
plt.show()
print(sample)
```

```python
F = sp.fft.fft(sample)
cal_mag = np.abs(F) * 2.0 / len(n)
plt.figure(figsize=(12, 9))
plt.plot(n, cal_mag, 'o')
plt.vlines(n, 0.0, cal_mag)
```
